```python
# ============================================================
# CAPSTONE PROJECT — Time Series Analysis 2026
# TASK 4: Volatility and Trend Analysis
# Consulting & Analytics Club, IIT Guwahati
# ============================================================
#
# What we do here:
#   1. Compute log returns for each stock
#   2. Estimate volatility using Rolling Std Dev
#   3. Model time-varying volatility using GARCH(1,1)
#   4. Apply STL decomposition (Trend + Seasonal + Residual)
#   5. Evaluate trend strength and direction
#   6. Summarize risk profiles for Task 5 portfolio allocation
# ============================================================


# ── STEP 0: Install dependencies (run this first in Colab)
# !pip install yfinance statsmodels arch scikit-learn --quiet


# ── STEP 1: Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import warnings
warnings.filterwarnings('ignore')

from statsmodels.tsa.seasonal import STL
from arch import arch_model   # For GARCH volatility modeling


# ── STEP 2: Define stocks
STOCKS = {
    "HDFCBANK.NS":   "Banking",
    "INFY.NS":       "Information Technology",
    "SUNPHARMA.NS":  "Pharmaceuticals",
    "HINDUNILVR.NS": "FMCG",
    "M&M.NS":        "Automobile"
}
colors = ['#1f77b4', '#9467bd', '#2ca02c', '#ff7f0e', '#d62728']

print("=" * 60)
print("TASK 4: Volatility and Trend Analysis")
print("=" * 60)


# ── STEP 3: Load data from Task 2
try:
    train_data = pd.read_csv("train_data.csv", index_col=0, parse_dates=True)
    test_data  = pd.read_csv("test_data.csv",  index_col=0, parse_dates=True)
    log_returns = pd.read_csv("log_returns.csv", index_col=0, parse_dates=True)
    close_prices = pd.concat([train_data, test_data])
    print("✓ Loaded data from Task 2 CSV files")
except FileNotFoundError:
    print("CSV files not found — re-downloading...")
    import yfinance as yf
    close_prices = yf.download(
        tickers=list(STOCKS.keys()),
        start="2021-01-01", end="2025-12-31",
        interval="1d", auto_adjust=True
    )
    if isinstance(close_prices.columns, pd.MultiIndex):
        close_prices = close_prices["Close"]
    close_prices = close_prices.ffill().bfill()
    train_data   = close_prices.loc[:"2025-06-30"]
    test_data    = close_prices.loc["2025-07-01":]
    log_returns  = np.log(close_prices / close_prices.shift(1)).dropna()
    print("✓ Data re-downloaded")

print(f"Full data: {close_prices.index[0].date()} → {close_prices.index[-1].date()}")


# ============================================================
# STEP 4: Log Returns Analysis
#
# Log return = ln(today's price / yesterday's price)
# This tells us the daily % change in a stock's value.
#
# Key statistics we look at:
#   Mean   : Average daily return (positive = stock grew overall)
#   Std Dev: How spread out the returns are (higher = more risky)
#   Min/Max: Worst and best single days
# ============================================================

print("\n── Step 4: Log Returns Analysis ──")

# Recompute log returns from full price series to be safe
log_returns = np.log(close_prices / close_prices.shift(1)).dropna()

print("\nLog Returns Summary Statistics:")
print("=" * 65)
stats = log_returns.describe()
print(stats.round(4))

# Annualized volatility = daily std dev × sqrt(252)
# 252 = number of trading days in a year
# This converts daily risk into a yearly risk measure
annual_vol = log_returns.std() * np.sqrt(252) * 100
print("\nAnnualized Volatility (%):")
for ticker, vol in annual_vol.items():
    risk_level = "Low" if vol < 20 else ("Medium" if vol < 35 else "High")
    print(f"  {ticker:<18} {vol:.2f}%  → {risk_level} risk")


# ── Plot Log Returns Distribution
fig, axes = plt.subplots(1, 5, figsize=(18, 4))
fig.suptitle("Log Returns Distribution — All Stocks", fontsize=13, fontweight='bold')

for idx, (ticker, sector) in enumerate(STOCKS.items()):
    ax = axes[idx]
    returns = log_returns[ticker].dropna()

    ax.hist(returns, bins=60, color=colors[idx], alpha=0.7, edgecolor='white', linewidth=0.3)
    ax.axvline(returns.mean(), color='black', linewidth=1.5, linestyle='--', label=f'Mean: {returns.mean():.4f}')
    ax.axvline(0, color='gray', linewidth=0.8, linestyle=':')
    ax.set_title(f"{ticker}\n({sector})", fontsize=9, fontweight='bold')
    ax.set_xlabel("Log Return", fontsize=8)
    ax.set_ylabel("Frequency", fontsize=8)
    ax.legend(fontsize=7)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig("task4_log_returns_distribution.png", dpi=150, bbox_inches='tight')
plt.show()
print("✓ Saved: task4_log_returns_distribution.png")


# ============================================================
# STEP 5: Rolling Standard Deviation (Volatility Over Time)
#
# Instead of one fixed volatility number, rolling std dev
# shows how volatility CHANGES over time.
#
# High rolling std dev period = market was turbulent
# Low rolling std dev period  = market was calm
#
# We use a 30-day window (1.5 months of trading days)
# ============================================================

print("\n── Step 5: Rolling Volatility (30-day window) ──")

WINDOW = 30
rolling_vol = log_returns.rolling(window=WINDOW).std() * np.sqrt(252) * 100

print("\nAverage Annualized Rolling Volatility (%):")
print(rolling_vol.mean().round(2))

# Plot rolling volatility
fig, axes = plt.subplots(5, 1, figsize=(14, 18), sharex=True)
fig.suptitle("Rolling 30-Day Annualized Volatility — All Stocks",
             fontsize=14, fontweight='bold')

for idx, (ticker, sector) in enumerate(STOCKS.items()):
    ax = axes[idx]
    vol = rolling_vol[ticker].dropna()

    ax.plot(vol.index, vol.values, color=colors[idx], linewidth=1.2)
    ax.fill_between(vol.index, vol.values, alpha=0.15, color=colors[idx])
    ax.axhline(vol.mean(), color='gray', linestyle='--', linewidth=0.8,
               label=f'Avg: {vol.mean():.1f}%')

    # Shade the test period (Jul–Dec 2025)
    ax.axvspan(pd.Timestamp("2025-07-01"), pd.Timestamp("2025-12-31"),
               alpha=0.08, color='red', label='Test period')

    ax.set_ylabel("Ann. Vol (%)", fontsize=9)
    ax.set_title(f"{ticker} ({sector})", fontsize=10, fontweight='bold')
    ax.legend(fontsize=8, loc='upper right')
    ax.grid(True, alpha=0.3)

axes[-1].set_xlabel("Date")
plt.tight_layout()
plt.savefig("task4_rolling_volatility.png", dpi=150, bbox_inches='tight')
plt.show()
print("✓ Saved: task4_rolling_volatility.png")


# ============================================================
# STEP 6: GARCH(1,1) Volatility Modeling
#
# Rolling std dev treats all past days equally.
# GARCH (Generalized AutoRegressive Conditional Heteroskedasticity)
# is smarter — it gives MORE weight to RECENT volatility.
#
# GARCH(1,1) means:
#   - Use 1 lag of past squared returns (ARCH term)
#   - Use 1 lag of past variance (GARCH term)
#
# In plain English:
#   "If the market was volatile yesterday, it's likely to be
#    volatile today too" — this is called volatility clustering.
#
# Output: conditional volatility — how risky is each stock TODAY?
# ============================================================

print("\n── Step 6: GARCH(1,1) Volatility Modeling ──")

garch_results   = {}
garch_forecasts = {}

fig, axes = plt.subplots(5, 1, figsize=(14, 18), sharex=True)
fig.suptitle("GARCH(1,1) Conditional Volatility — All Stocks",
             fontsize=14, fontweight='bold')

for idx, (ticker, sector) in enumerate(STOCKS.items()):
    print(f"\n  Fitting GARCH(1,1) for {ticker}...")

    # GARCH models work on returns scaled by 100 (percentage returns)
    returns = log_returns[ticker].dropna() * 100

    # Fit GARCH(1,1) model
    # vol='Garch' → use GARCH volatility model
    # p=1 → 1 ARCH lag (recent shock effect)
    # q=1 → 1 GARCH lag (persistence of volatility)
    model  = arch_model(returns, vol='Garch', p=1, q=1, dist='normal')
    result = model.fit(disp='off')  # disp='off' suppresses verbose output

    # Extract conditional volatility (annualized)
    cond_vol = result.conditional_volatility * np.sqrt(252)

    # Forecast next 5 days volatility
    forecast    = result.forecast(horizon=5, reindex=False)
    vol_forecast = np.sqrt(forecast.variance.values[-1]) * np.sqrt(252)

    garch_results[ticker]   = result
    garch_forecasts[ticker] = {
        "Current Vol (%)":      round(float(cond_vol.iloc[-1]), 2),
        "5-day Forecast Vol (%)": round(float(vol_forecast.mean()), 2),
        "Alpha (ARCH)":         round(float(result.params.get('alpha[1]', 0)), 4),
        "Beta (GARCH)":         round(float(result.params.get('beta[1]', 0)), 4),
        "Persistence":          round(float(result.params.get('alpha[1]', 0) +
                                result.params.get('beta[1]', 0)), 4)
    }

    print(f"    Current volatility  : {garch_forecasts[ticker]['Current Vol (%)']:.2f}%")
    print(f"    5-day forecast vol  : {garch_forecasts[ticker]['5-day Forecast Vol (%)']:.2f}%")
    print(f"    Persistence (α+β)   : {garch_forecasts[ticker]['Persistence']:.4f}")

    # Plot
    ax = axes[idx]
    ax.plot(cond_vol.index, cond_vol.values, color=colors[idx], linewidth=1.0)
    ax.fill_between(cond_vol.index, cond_vol.values, alpha=0.15, color=colors[idx])
    ax.axhline(cond_vol.mean(), color='gray', linestyle='--', linewidth=0.8,
               label=f"Avg: {cond_vol.mean():.1f}%")
    ax.set_title(f"{ticker} ({sector})  |  "
                 f"Current: {garch_forecasts[ticker]['Current Vol (%)']:.1f}%  |  "
                 f"5-day forecast: {garch_forecasts[ticker]['5-day Forecast Vol (%)']:.1f}%",
                 fontsize=9, fontweight='bold')
    ax.set_ylabel("Ann. Vol (%)", fontsize=9)
    ax.legend(fontsize=8)
    ax.grid(True, alpha=0.3)

axes[-1].set_xlabel("Date")
plt.tight_layout()
plt.savefig("task4_garch_volatility.png", dpi=150, bbox_inches='tight')
plt.show()
print("\n✓ Saved: task4_garch_volatility.png")

# Print GARCH summary
print("\nGARCH(1,1) Summary:")
garch_df = pd.DataFrame(garch_forecasts).T
print(garch_df.to_string())
garch_df.to_csv("task4_garch_results.csv")
print("✓ Saved: task4_garch_results.csv")


# ============================================================
# STEP 7: STL Decomposition (Trend + Seasonal + Residual)
#
# STL = Seasonal and Trend decomposition using Loess
# It splits each stock's price into 3 clean components:
#
#   Trend    → The overall long-term direction
#   Seasonal → Repeating yearly pattern
#   Residual → What's left after removing trend and seasonal
#              (essentially random noise)
#
# Why this matters for Task 5:
#   - Strong upward trend = good candidate for allocation
#   - High residual = unpredictable stock = allocate less
# ============================================================

print("\n── Step 7: STL Decomposition ──")

stl_results = {}

fig, axes = plt.subplots(5, 4, figsize=(20, 22))
fig.suptitle("STL Decomposition — Trend + Seasonal + Residual",
             fontsize=14, fontweight='bold')

col_labels = ["Observed\n(Raw Price)", "Trend", "Seasonal", "Residual\n(Noise)"]

for idx, (ticker, sector) in enumerate(STOCKS.items()):
    series = close_prices[ticker].dropna()

    # STL decomposition
    # period=252 → one full year of trading days
    # seasonal=13 → seasonal smoother window (must be odd)
    stl    = STL(series, period=252, seasonal=13)
    result = stl.fit()

    # Extract components
    trend    = result.trend
    seasonal = result.seasonal
    residual = result.resid

    # Trend direction: compare last 6 months vs first 6 months of trend
    recent_trend = trend.iloc[-126:].mean()   # Last 6 months
    earlier_trend= trend.iloc[:126].mean()    # First 6 months
    trend_change = ((recent_trend - earlier_trend) / earlier_trend) * 100

    # Trend strength = how much of the variation is explained by trend
    # Higher = stronger trend (stock follows a clear direction)
    trend_strength = 1 - (residual.var() / (trend + residual).var())
    trend_strength = max(0, min(1, trend_strength))  # Clamp between 0 and 1

    direction = "Upward ↑" if trend.iloc[-1] > trend.iloc[len(trend)//2] else "Downward ↓"

    stl_results[ticker] = {
        "Trend Direction":    direction,
        "Trend Change (%)":   round(trend_change, 2),
        "Trend Strength":     round(trend_strength, 4),
        "Recent Trend (₹)":  round(float(recent_trend), 2),
        "Seasonal Range (₹)": round(float(seasonal.max() - seasonal.min()), 2),
        "Residual Std (₹)":  round(float(residual.std()), 2)
    }

    print(f"\n  {ticker} ({sector})")
    print(f"    Direction     : {direction}")
    print(f"    Trend change  : {trend_change:+.2f}%")
    print(f"    Trend strength: {trend_strength:.4f} (0=none, 1=perfect)")
    print(f"    Seasonal range: ₹{stl_results[ticker]['Seasonal Range (₹)']}")

    # Plot all 4 components
    components = [series, trend, seasonal, residual]
    comp_colors = [colors[idx], colors[idx], 'orange', 'gray']

    for col, (comp, clr, label) in enumerate(zip(components, comp_colors, col_labels)):
        ax = axes[idx][col]
        ax.plot(comp.index, comp.values, color=clr, linewidth=0.8)
        if col == 3:  # Residual — add zero line
            ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
        if idx == 0:
            ax.set_title(label, fontsize=10, fontweight='bold')
        if col == 0:
            ax.set_ylabel(f"{ticker}\n({sector})", fontsize=8)
        ax.grid(True, alpha=0.3)
        ax.tick_params(axis='x', labelsize=7)
        ax.tick_params(axis='y', labelsize=7)

plt.tight_layout()
plt.savefig("task4_stl_decomposition.png", dpi=150, bbox_inches='tight')
plt.show()
print("\n✓ Saved: task4_stl_decomposition.png")

# Save STL results
stl_df = pd.DataFrame(stl_results).T
print("\nSTL Decomposition Summary:")
print(stl_df.to_string())
stl_df.to_csv("task4_stl_results.csv")
print("✓ Saved: task4_stl_results.csv")


# ============================================================
# STEP 8: Correlation Matrix
#
# Correlation measures how similarly two stocks move together:
#   +1.0 = move perfectly together (bad for diversification)
#    0.0 = no relationship
#   -1.0 = move in opposite directions (great for diversification)
#
# We want stocks with LOW correlation to reduce portfolio risk.
# If one stock falls, a low-correlated stock might stay flat or rise.
# ============================================================

print("\n── Step 8: Correlation Matrix ──")

# Use log returns for correlation (standard in finance)
corr_matrix = log_returns.corr()

# Rename columns for cleaner display
short_names = {t: t.replace(".NS", "") for t in STOCKS.keys()}
corr_display = corr_matrix.rename(index=short_names, columns=short_names)

print("\nReturn Correlation Matrix:")
print(corr_display.round(3).to_string())

# Plot correlation heatmap
fig, ax = plt.subplots(figsize=(8, 6))

im = ax.imshow(corr_display.values, cmap='RdYlGn', vmin=-1, vmax=1, aspect='auto')
plt.colorbar(im, ax=ax, label='Correlation')

tickers_short = list(short_names.values())
ax.set_xticks(range(len(tickers_short)))
ax.set_yticks(range(len(tickers_short)))
ax.set_xticklabels(tickers_short, rotation=45, ha='right', fontsize=10)
ax.set_yticklabels(tickers_short, fontsize=10)

# Add correlation values as text inside each cell
for i in range(len(tickers_short)):
    for j in range(len(tickers_short)):
        val = corr_display.values[i][j]
        color = 'white' if abs(val) > 0.7 else 'black'
        ax.text(j, i, f"{val:.2f}", ha='center', va='center',
                fontsize=11, fontweight='bold', color=color)

ax.set_title("Return Correlation Matrix — All 5 Stocks\n(Green = Low correlation = Good diversification)",
             fontsize=11, fontweight='bold')
plt.tight_layout()
plt.savefig("task4_correlation_heatmap.png", dpi=150, bbox_inches='tight')
plt.show()
print("✓ Saved: task4_correlation_heatmap.png")


# ============================================================
# STEP 9: Final Risk Profile Summary
# This feeds directly into Task 5 portfolio allocation
# ============================================================

print("\n── Step 9: Final Risk Profile Summary for Task 5 ──")

risk_profile = {}
for ticker, sector in STOCKS.items():
    ann_vol    = annual_vol[ticker]
    garch_vol  = garch_forecasts[ticker]["5-day Forecast Vol (%)"]
    trend_dir  = stl_results[ticker]["Trend Direction"]
    trend_str  = stl_results[ticker]["Trend Strength"]
    resid_std  = stl_results[ticker]["Residual Std (₹)"]

    # Risk score: combine annualized vol and GARCH forecast
    # Lower = safer stock
    risk_score = round((ann_vol + garch_vol) / 2, 2)

    # Inverse volatility weight (for Task 5 Strategy B)
    # Stocks with lower volatility get higher weight
    inv_vol    = 1 / risk_score

    risk_profile[ticker] = {
        "Sector":             sector,
        "Ann. Vol (%)":       round(ann_vol, 2),
        "GARCH 5-day Vol (%)": garch_vol,
        "Trend Direction":    trend_dir,
        "Trend Strength":     trend_str,
        "Residual Std (Rs)":  resid_std,
        "Risk Score":         risk_score,
        "Inv. Vol Weight":    round(inv_vol, 6)
    }

risk_df = pd.DataFrame(risk_profile).T

# Normalize inverse volatility weights to sum to 1
total_inv_vol = risk_df["Inv. Vol Weight"].sum()
risk_df["Normalized Weight (%)"] = (risk_df["Inv. Vol Weight"] / total_inv_vol * 100).round(2)

print("\nFull Risk Profile:")
print(risk_df.to_string())
risk_df.to_csv("task4_risk_profile.csv")
print("\n✓ Saved: task4_risk_profile.csv")

print("\nVolatility-Based Portfolio Weights (Strategy B for Task 5):")
print(f"{'Stock':<18} {'Risk Score':>12} {'Weight (%)':>12}")
print("-" * 45)
for ticker in STOCKS.keys():
    print(f"{ticker:<18} {risk_df.loc[ticker,'Risk Score']:>12.2f} "
          f"{risk_df.loc[ticker,'Normalized Weight (%)']:>11.2f}%")


# ── FINAL SUMMARY
print("\n" + "=" * 60)
print("TASK 4 COMPLETE ✓")
print("=" * 60)
print("\nKey Findings:")
print("\n  Volatility (riskiest → safest):")
sorted_vol = annual_vol.sort_values(ascending=False)
for ticker, vol in sorted_vol.items():
    print(f"    {ticker:<18} {vol:.2f}% annualized")

print("\n  Trend Direction:")
for ticker, info in stl_results.items():
    print(f"    {ticker:<18} {info['Trend Direction']}  "
          f"(strength: {info['Trend Strength']:.2f})")

print("\n  Correlation insight:")
corr_vals = corr_matrix.values
np.fill_diagonal(corr_vals, np.nan)
low_corr  = [(corr_display.index[i], corr_display.columns[j], round(corr_vals[i][j], 3))
             for i in range(5) for j in range(i+1, 5)
             if not np.isnan(corr_vals[i][j]) and corr_vals[i][j] < 0.5]
print(f"    Pairs with correlation < 0.5 (good diversification):")
for a, b, c in low_corr:
    print(f"      {a} & {b}: {c}")

print("\nFiles saved:")
print("  task4_log_returns_distribution.png")
print("  task4_rolling_volatility.png")
print("  task4_garch_volatility.png")
print("  task4_stl_decomposition.png")
print("  task4_correlation_heatmap.png")
print("  task4_garch_results.csv")
print("  task4_stl_results.csv")
print("  task4_risk_profile.csv")
print("\nNext → Task 5: Portfolio Construction & Capital Allocation")
print("=" * 60)
```
